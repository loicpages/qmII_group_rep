```
#importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
```

```
#importing data with pandas 
data_path = '/content/df_final_use.csv'
general_df = pd.read_csv(data_path, encoding ='latin-1', sep=';')
general_df.head()
```

```
#creating a function to create a graph showing the evolution of food availability, food production, food imports, food exports, and food aid over time for a specific country 
def plot_country(country):
  country_df = general_df[general_df['Country']==country]
  
  #plotting multiple garphs on the same axis : creating and ax object to then use it as an argument  
  ax = country_df.plot(x='Year', y='Food Availibility per capita (kg/cap/year)', kind='line', c = 'red', label = 'Food Availibility per capita (kg/cap/year)')
  country_df.plot(x='Year', y= 'Production Quantity per capita (kg/cap/year)', kind='line', c = 'black', ax=ax, label = 'Production Quantity per capita (kg/cap/year)')
  country_df.plot(x='Year', y= 'Imports (all) per capita (kg/cap/year)', kind='line', c = 'green', ax=ax, label = 'Imports (all) per capita (kg/cap/year)')
  country_df.plot(x='Year', y= 'Exports (all) per capita (kg/cap/year)', kind='line', c = 'yellow', ax=ax, label = 'Exports (all) per capita (kg/cap/year)')
  country_df.plot(x='Year', y= 'Total Food Aid', kind='line', c = 'pink', ax=ax, label = 'Total Food Aid per capita (kg/cap/year)')

  ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
  plt.title(country)
  plt.ylabel('Quantity per capita (kg/cap/year)')

#calling the function for every country in Southern Africa 
plot_country('Zimbabwe')
plot_country('Zambia')
plot_country('Swaziland')
plot_country('Namibia')
plot_country('Mozambique')
plot_country('Malawi')
plot_country('Madagascar')
plot_country('Lesotho')


```

```
#Importing a few liraries
import pandas as pd
import plotly.express as px
from functools import reduce
```

```
#importing our data frame
data_path = '/content/df_final_use.csv' 
frame = pd.read_csv(data_path)
frame
```

```
#Alright, so let's create a function that will run through this frame and average all food availibility values per region each year.
#We name this function Average and pass on two arguments to it: the region we want to average and the name of the data frame.
#It is important that all the data frames have different names so that we can concatenate them together at a later stage!

def Average(region):

  region_df = frame[frame['Region'] == region]
  region_food_df = region_df.groupby('Year')['Food Availibility per capita (kg/cap/year)'].mean()
  column_names = ["Year", "Food Availibility per capita (kg/cap/year)", "Region"]
  df_name = pd.DataFrame(columns = column_names)
  df_name['Year'] = region_food_df.index.to_list()
  df_name['Food Availibility per capita (kg/cap/year)'] = region_food_df.to_list()
  df_name['Region'] = region

  return df_name
```

```
#We now use the average function for each region as follows:
df1 = Average('Southern Africa')
df2 = Average('North Africa')
df3 = Average('Central Africa')
df4 = Average('Caucasus and Middle East')
df5 = Average('South Asia')
df6 = Average('West Africa')
df7 = Average('South America')
df8 = Average('East Africa')
df9 = Average('South East Asia')
df10 = Average('Central and East Asia')
df11 = Average('Europe')
```

```
#Concatenating all of the data_frames into one:
frames = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11]
all_frames = pd.concat(frames)
all_frames
```

```
fig = px.line(all_frames, x = 'Year', y = 'Food Availibility per capita (kg/cap/year)',
              color = 'Region', line_group = 'Region',
              labels={"Food Availibility per capita (kg/cap/year)": "Food Availibility per capita (kg/cap/year)", "Region": "Region"},
              title = 'Countries: Food Availibility per capita (kg/cap/year)',
              hover_name = 'Region',
              )

fig.for_each_trace(lambda t: t.update(name=t.name.replace("Region=", "")))
fig.show()
```
